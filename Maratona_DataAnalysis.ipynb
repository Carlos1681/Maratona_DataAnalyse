```python
import pandas as pd
import seaborn as sns
from google.colab import files
import io
uploaded = files.upload()
```

```python
# Ler os dados do arquivo.
df = pd.read_csv(io.BytesIO(uploaded['TWO_CENTURIES_OF_UM_RACES (1).csv']))
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
#Limpeza de dados
#Corridas de 50km e 50mi
#Ano 2019
df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Year of event'] == 2019)]
```

```python
#Eventos no Brasil
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'BRA']
```

```python
#Combinar os dois filtros
df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Year of event'] == 2019) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'BRA')]
```

```python
new_df = df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Year of event'] == 2019) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'BRA')]
```

```python
new_df.head()
```

```python
new_df.shape
```

```python
#Alterar o nome do evento, removendo a sigla do país
new_df['Event name'].str.split('(').str.get(0)
```

```python
new_df['Event name'] = new_df['Event name'].str.split('(').str.get(0)
```

```python
new_df.head()
```

```python
#Cálculo de idade
new_df['Athlete age'] = 2019 - new_df['Athlete year of birth']
```

```python
new_df['Athlete performance'] = new_df['Athlete performance'].str.split(' ').str.get(0)
```

```python
new_df.head()
```

```python
#Removendo colunas não utilizadas
new_df = new_df.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
new_df.head()
```

```python
#Removendo valores nulos
new_df.isna().sum()
```

```python
new_df[new_df['Athlete age'].isna() == 1]
```

```python
new_df = new_df.dropna()
```

```python
#Checando por valores duplicados
new_df[new_df.duplicated() == True]
```

```python
#Resetando o index das colunas, para começar do 0
new_df = new_df.reset_index(drop=True)
```

```python
#Alterando o tipo de algumas colunas
new_df['Athlete age'] = new_df['Athlete age'].astype(int)
```

```python
new_df['Athlete average speed'] = new_df['Athlete average speed'].astype(float)
```

```python
new_df.dtypes
```

```python
#Renomeando colunas
new_df = new_df.rename(columns = {
    'Year of event': 'ano',
    'Event dates': 'data_corrida',
    'Event name': 'nome_corrida',
    'Event distance/length': 'distancia',
    'Event number of finishers': 'corrida_quantidade_finalizaram',
    'Athlete performance': 'performance',
    'Athlete gender': 'genero',
    'Athlete average speed': 'velocidade_media',
    'Athlete ID': 'id',
    'Athlete age': 'idade'
})
```

```python
new_df.head()
```

```python
#Reordenar colunas
df_new = new_df[['data_corrida','nome_corrida','distancia','corrida_quantidade_finalizaram','id','genero','idade','performance','velocidade_media']]
```

```python
df_new.head()
```

```python
df_new[df_new['nome_corrida'] == "X-Treme Run Gramado 50 Kms "]
```

```python
#Gráficos
sns.histplot(df_new['distancia'])
```

```python
df_new['distancia'].value_counts() #Houve apenas maratonas de 50km no Brasil durante 2019, segundo os dados analisados.
```

```python
sns.histplot(df_new, x = 'distancia', hue = 'genero')
```

```python
sns.displot(df_new[df_new['distancia'] == '50km']['velocidade_media'])
```

```python
sns.violinplot(data = df_new, x = 'distancia', y = 'velocidade_media', hue = 'genero', split = True, inner = 'quart', linewidth= 1)
```

```python
sns.lmplot(data = df_new, x = 'idade', y = 'velocidade_media', hue = 'genero')
```

```python
#Obtendo informações através do dados.
```

```python
#Velocidade média dos corredores.
df_new.groupby('genero')['velocidade_media'].mean()
```

```python
#Separando por idade dos competidores, para verificar a velocidade media e quantidade de competidores por idade.
#Considerando idades com pelo menos 10 corredores.
#Maiores velocidade médias
df_new.groupby('idade')['velocidade_media'].agg(['mean','count']).sort_values('mean', ascending=False).query('count>9').head()
```

```python
#Menores velocidade médias
df_new.groupby('idade')['velocidade_media'].agg(['mean','count']).sort_values('mean', ascending=True).query('count>9').head()
```

```python
#Análise da performance com base na estação do ano
df_new['mes_corrida'] = df_new['data_corrida'].str.split('.').str.get(1).astype(int)
```

```python
df_new['estacao_ano'] = df_new['mes_corrida'].apply(lambda x: 'Verão' if x > 11 else 'Primavera' if x > 8 else 'Inverno' if x > 5 else 'Outono' if x > 2 else 'Verão')
```

```python
df_new.head(25)
```

```python
#Velocidade média por estação do ano
df_new.groupby('estacao_ano')['velocidade_media'].agg(['mean','count']).sort_values('mean', ascending = False)
```
