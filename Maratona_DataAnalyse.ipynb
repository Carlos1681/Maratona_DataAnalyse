```python
import pandas as pd
import seaborn as sns
import io
from google.colab import files
uploaded = files.upload()
```

```python
# Ler os dados do arquivo.
df = pd.read_csv(io.BytesIO(uploaded['TWO_CENTURIES_OF_UM_RACES.csv']))
```

```python
#Limpeza de dados
#Corridas de 50km e 50mi
#Ano 2019
df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Year of event'] == 2019)]
```

```python
#Eventos no Brazil
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'BRA']
```

```python
#Combinar os dois filtros
df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Year of event'] == 2019) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'BRA')]
```

```python
new_df = df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Year of event'] == 2019) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'BRA')]
```

```python
new_df.head()
```

```python
new_df.shape
```

```python
#Alterar o nome do evento, removendo a sigla do país
new_df['Event name'].str.split('(').str.get(0)
```

```python
new_df['Event name'] = new_df['Event name'].str.split('(').str.get(0)
```

```python
new_df.head()
```

```python
#Cálculo de idade
new_df['Athlete age'] = 2019 - new_df['Athlete year of birth']
```

```python
new_df['Athlete performance'] = new_df['Athlete performance'].str.split(' ').str.get(0)
```

```python
new_df.head()
```

```python
#Removendo colunas não utilizadas
new_df = new_df.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
new_df.head()
```

```python
#Removendo valores nulos
new_df.isna().sum()
```

```python
new_df[new_df['Athlete age'].isna() == 1]
```

```python
new_df = new_df.dropna()
```

```python
#Checando por valores duplicados
new_df[new_df.duplicated() == True]
```

```python
#Resetando o index das colunas, para começar do 0
new_df = new_df.reset_index(drop=True)
```

```python
#Alterando o tipo de algumas colunas
new_df['Athlete age'] = new_df['Athlete age'].astype(int)
```

```python
new_df['Athlete average speed'] = new_df['Athlete average speed'].astype(float)
```

```python
new_df.dtypes
```

```python
#Renomeando colunas
new_df = new_df.rename(columns = {
    'Year of event': 'ano',
    'Event dates': 'data_corrida',
    'Event name': 'nome_corrida',
    'Event distance/length': 'distancia',
    'Event number of finishers': 'corrida_quantidade_finalizaram',
    'Athlete performance': 'performance',
    'Athlete gender': 'genero',
    'Athlete average speed': 'velocidade_media',
    'Athlete ID': 'id',
    'Athlete age': 'idade'
})
```
